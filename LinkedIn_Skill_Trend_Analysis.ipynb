```python
import matplotlib.pyplot as plt
import seaborn as sns

top_titles = df['job_title'].value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_titles.values, y=top_titles.index, palette='viridis')
plt.title("Top 10 Most In-Demand Job Titles")
plt.xlabel("Number of Job Postings")
plt.ylabel("Job Title")
plt.tight_layout()
plt.show()
```

```python
top_cities = df['search_city'].value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_cities.values, y=top_cities.index, palette='coolwarm')
plt.title("Top 10 Cities with Most LinkedIn Job Postings")
plt.xlabel("Number of Postings")
plt.ylabel("City")
plt.tight_layout()
plt.show()
```

```python
type_counts = df['job_type'].value_counts()

plt.figure(figsize=(6, 4))
sns.barplot(x=type_counts.index, y=type_counts.values, palette='pastel')
plt.title("Job Type Distribution")
plt.xlabel("Job Type")
plt.ylabel("Number of Postings")
plt.tight_layout()
plt.show()
```

```python
level_counts = df['job level'].value_counts()

plt.figure(figsize=(6, 4))
sns.barplot(x=level_counts.index, y=level_counts.values, palette='Set2')
plt.title("Job Level Distribution")
plt.xlabel("Job Level")
plt.ylabel("Number of Postings")
plt.tight_layout()
plt.show()
```

```python
ny_df = df[df['search_city'].str.lower() == 'new york']
ny_types = ny_df['job_type'].value_counts().head(5)

plt.figure(figsize=(6, 4))
sns.barplot(x=ny_types.index, y=ny_types.values, palette='Blues')
plt.title("Top Job Types in New York")
plt.xlabel("Job Type")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
file_path = "C:/Users/aishw/OneDrive/Documents/postings.csv"
df = pd.read_csv(file_path)
```

```python
df.columns = df.columns.str.lower().str.strip()
```

```python
df.head(), df.columns
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a pivot table of job_title vs job_type
pivot = df.pivot_table(index='job_title', columns='job_type', aggfunc='size', fill_value=0)

# Show only top 10 job titles
pivot_top10 = pivot.loc[df['job_title'].value_counts().head(10).index]

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_top10, annot=True, fmt='d', cmap='Blues')
plt.title("Job Type Distribution Across Top 10 Job Titles")
plt.ylabel("Job Title")
plt.xlabel("Job Type")
plt.tight_layout()
plt.show()
```
